## Previo 

### Previo de la Práctica 1

Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- Definición de URDF [1](https://wiki.ros.org/urdf).

  URDF (Unified Robot Description Format) es un formato basado en XML utilizado para describir la estructura física de un robot en el sistema operativo de robots (ROS, por sus siglas en inglés). Un archivo URDF define componentes del robot como los enlaces (partes rígidas) y las articulaciones (joints), proporcionando la configuración geométrica, la inercia, la ubicación de los sensores, los actuadores y otros elementos de los componentes del robot. Estos archivos son esenciales para la simulación y visualización en herramientas como RVIZ y Gazebo.

- Tipos de juntas que están definidas dentro del formato URDF [2](https://wiki.ros.org/urdf/XML/joint).

    En URDF, las juntas o "joints" definen cómo los enlaces están conectados entre sí y cómo se pueden mover. Los tipos de juntas más comunes son:

          - Fixed: No permite movimiento entre los enlaces conectados. Es una conexión rígida.
          - Revolute: Permite rotación alrededor de un eje fijo dentro de un rango de ángulos.
          - Continuous: Es similar a la junta revolute, pero sin límite de ángulo, permitiendo una rotación completa.
          - Prismatic: Permite movimiento lineal a lo largo de un eje, dentro de un rango determinado.
          - Floating: Permite movimiento y rotación sin restricciones en los seis grados de libertad (tres traslaciones y tres rotaciones).
          - Planar: Permite movimiento en un plano bidimensional, permitiendo traslaciones en X e Y y rotación alrededor del    eje Z.

   
- Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS [3](https://wiki.ros.org/urdf/XML/link).

  Los elementos "primitivos" se refieren a las formas geométricas básicas utilizadas en URDF para modelar la estructura del robot. Estos son:

      - Box: Un cuboide definido por su longitud, anchura y altura.

      - Cylinder: Un cilindro definido por su radio y altura.

      - Sphere: Una esfera definida por su radio.

      - Mesh: Un modelo de malla tridimensional más complejo que puede ser importado desde un archivo STL o DAE.

      - Estos elementos permiten crear representaciones geométricas simples del robot para la simulación.

- Que es un nodo y que es un topico en ROS 2 [4](https://docs.ros.org/en/foxy/Tutorials/Beginner-Client-Libraries/Writing-A-Simple-Cpp-Publisher-And-Subscriber.html) [5](https://docs.ros.org/en/foxy/Tutorials/Beginner-CLI-Tools/Understanding-ROS2-Topics/Understanding-ROS2-Topics.html).

      - Nodo: En ROS 2, un nodo es una unidad de proceso independiente que realiza tareas específicas. Los nodos se comunican entre sí de manera asíncrona mediante tópicos, servicios o acciones. Por ejemplo, un nodo puede estar a cargo de controlar los motores de un robot, mientras que otro puede gestionar la captación de datos de los sensores.

      - Tópico: Un tópico en ROS 2 es un canal de comunicación utilizado para que los nodos intercambien mensajes de datos. Los nodos pueden publicar mensajes en un tópico (actuando como publicadores) o suscribirse a un tópico para recibir mensajes (actuando como suscriptores). Este mecanismo permite la comunicación entre nodos de manera flexible y eficiente.
  
- Que es RVIZ [6](https://wiki.ros.org/rviz).

  RVIZ es una herramienta de visualización 3D en ROS que permite a los usuarios ver el estado de un robot en tiempo real, incluyendo su estructura, sensores, trayectorias y entorno. Es útil para depurar y monitorear la simulación y el comportamiento de un robot. Los usuarios pueden agregar diferentes tipos de datos visuales, como mapas, imágenes de cámaras, datos de sensores láser, e incluso la posición y orientación del robot.

En caso de integrar imagenes, colocarlas en un carpeta definida como Imagenes_p1 y colocar la bibliografía al final de esta sección.

El desarrollo de la práctica se divide en dos partes:

1. La primera parte de realizará en clase, en la cual se desarrollará y simulará de descripción de un eslabón cayo movimiento está restringido por una junta rotacional.
   
***Bloque de código***

Código de dos juntas:

```js
<?xml version="1.0" ?>

<robot name="basic_link">
    
    <link name="base_link">
            <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.1 0.12 0.15"/>
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="red">
                <color rgba="1.0 0.0 0.0 1.0"/>
            </material>
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.0 0.0 0.1" rpy="0 0 0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.1 0.15 0.2"/>
            </geometry>
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="1.0"/>
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0 0 0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="1.0" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/>
        </inertial>

    </link>
    
    <link name="link_1">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientación de origen del eslabon-->
            <origin xyz="0.0 0.0 0.25" rpy="0 0 0" />
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.05 0.05 0.5" />
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="green">
                <color rgba="1.0 0.0 0.0 1.0" />
            </material>
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.0 0.0 0.0" rpy="0 0 0" />
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.15 0.1 0.2" />
            </geometry>
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="1.0" />
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0 0 0 " />
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="1.0" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0" />
        </inertial>
    </link>

    <joint name="base_link_to_link_1_joint" type="revolute">
        <parent link="base_link"/>
        <child link="link_1"/>
        <axis xyz="1 0 0"/>
        <limit lower="-3.14159" upper= "3.14159" velocity="50.0" effort="1000.0"/>
        <origin xyz="0 0 0.15" rpy="0 0 0"/>
    </joint>
</robot>
```



## Conclusiones 

La práctica de modificar el archivo URDF para agregar un nuevo eslabón y conectarlo mediante una junta de revolución permitió comprender de manera más profunda cómo se estructuran y describen los robots en el entorno de ROS. Al implementar una nueva junta, se demostró la versatilidad del formato URDF para modelar robots con diversas configuraciones cinemáticas. Además, al visualizar la estructura en herramientas como RViz o Gazebo, se facilitó la validación de la correcta configuración de las juntas y los enlaces.

Este tipo de ejercicios es fundamental en el desarrollo de robots, ya que permite iterar rápidamente sobre el diseño y verificar su funcionalidad antes de pasar a la implementación física. La capacidad de simular el comportamiento del robot es crucial para el análisis y optimización de su rendimiento, lo cual es esencial en la ingeniería robótica y mecatrónica.

Finalmente, la práctica refuerza el entendimiento de conceptos clave, como los tipos de juntas (revolución en este caso), la interconexión de los enlaces y la representación geométrica simplificada del robot. Esto forma una base sólida para proyectos más complejos, donde la simulación y el control de robots desempeñan un papel crucial en la planificación y ejecución de tareas en entornos reales y simulados.
